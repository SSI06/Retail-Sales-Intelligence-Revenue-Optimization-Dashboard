```python
import pandas as pd
```

```python
# ---------------- LOAD DATA ----------------
df = pd.read_csv("sales_data_big.csv")
```

```python
print("Initial Shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
```

```python
# ---------------- CHECK NULL VALUES ----------------
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# ---------------- CHECK DATA TYPES ----------------
print("\nData Types:")
print(df.dtypes)
```

```python
# ---------------- CONVERT DATE ----------------
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
# ---------------- CHECK DUPLICATES ----------------
duplicates = df.duplicated().sum()
print("\n Duplicate Rows:", duplicates)
```

```python
# ---------------- BASIC STATS ----------------
print("\nNumerical Summary:")
print(df.describe())
```

```python
# ---------------- UNIQUE VALUES ----------------
print("\nUnique Regions:", df['region'].unique())
print("Unique Categories:", df['category'].unique())
print("Unique Products:", df['product'].unique())
```

```python
# ---------------- SAVE CLEANED DATA ----------------
df.to_csv("cleaned_sales_data.csv", index=False)

print("\nData Cleaning Completed & Saved!")
```
